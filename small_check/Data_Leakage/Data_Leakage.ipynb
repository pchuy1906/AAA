```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the data
data = pd.read_csv('AER_credit_card_data.csv')
data.head()
```

```python
data.dtypes
```

```python
col_float = list(data.select_dtypes(include=['float64']).columns)
print (col_float)
```

```python
fig, axs = plt.subplots(1, len(col_float), figsize=(16,6))
axes = axs.flatten()
labels = ["card = Yes","card = No"]

num = 0
for feature in col_float:
    data_hist = [data[data['card']=='yes'][feature], data[data['card']=='no'][feature]]
    axes[num].hist(data_hist, label=labels)
    axes[num].legend(prop={'size': 10})
    axes[num].set_title(feature)
    num += 1
```

```python
col_int = list(data.select_dtypes(include=['int64']).columns)
print (col_int)
```

```python
fig, axs = plt.subplots(1, len(col_int), figsize=(16,6))
axes = axs.flatten()
labels = ["card = Yes","card = No"]

num = 0
for feature in col_int:
    data_hist = [data[data['card']=='yes'][feature], data[data['card']=='no'][feature]]
    axes[num].hist(data_hist, label=labels)
    axes[num].legend(prop={'size': 10})
    axes[num].set_title(feature)
    num += 1
```

```python
col_object = list(data.select_dtypes(include=['object']).columns)
print (col_object)
```

```python
fig, axs = plt.subplots(1, len(col_object), figsize=(16,6))
axes = axs.flatten()
labels = ["card = Yes","card = No"]

num = 0
for feature in col_object:
    data_hist = [data[data['card']=='yes'][feature], data[data['card']=='no'][feature]]
    axes[num].hist(data_hist, label=labels)
    axes[num].legend(prop={'size': 10})
    axes[num].set_title(feature)
    num += 1
```

```python
import seaborn as sns

corr = data[col_num].corr()

plt.figure(figsize = (10,8))

sns.heatmap(corr, xticklabels = corr.columns, yticklabels = corr.columns, cmap="YlGnBu",annot=False)

plt.show()
```
