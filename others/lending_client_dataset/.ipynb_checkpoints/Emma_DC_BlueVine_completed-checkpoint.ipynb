### Duplicates

There is no duplicate in the dataset. But there are duplicates in `client_id` - might be returning clients.

```python
data[data.duplicated(keep=False)].index
```

```python
data[data.duplicated(subset=['client_id'],keep=False)].head()
```

```python
data_orig['client_id'].value_counts().head()
```

```python
numerical_features = data.dtypes[data.dtypes != 'object'].index.values
plt.figure(1, figsize = (20,15))
count = 0
for feature in numerical_features[2:12]:
    plt.subplot(str(33) + str(count))
    count += 1
    data[feature].hist()
    plt.title('Frequency distribution of '+ feature)
    plt.xlabel(feature, fontsize=12)
    plt.xticks(rotation=45)
    #plt.show()
```

```python

plt.figure(2, figsize = (20,5))
count = 0
for feature in numerical_features[12:]:
    plt.subplot(str(13) + str(count))
    count += 1
    data[feature].hist()
    plt.title('Frequency distribution of '+ feature)
    plt.xlabel(feature, fontsize=12)
    plt.xticks(rotation=45)
```
